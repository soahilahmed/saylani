1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance
is insufficient.

```python
class BankAccount:
    def __init__(self, holder_name, account_number, balance):
        self.holder_name = holder_name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")


# Object create
acc1 = BankAccount("Ali", "12345", 1000)

acc1.deposit(500)      # Deposited 500. New balance: 1500
acc1.withdraw(2000)    # Insufficient balance!
acc1.withdraw(1000)    # Withdrew 1000. New balance: 500lf): 

    
```

2. Design a class 'Student' with attributes name and marks. Create a method to calculate grade
based on marks using conditions (A, B, C, Fail).

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "Fail"


s1 = Student("sohail", 95)
s2 = Student("tuha faiz", 80)
s3 = Student("muhammad furqan", 30)

print(s1.name, s1.grade())  
print(s2.name, s2.grade())  
print(s3.name, s3.grade()) 
```

3. Create a class 'ElectricityBill' for a KE-electric consumer. Include units consumed and calculate
bill based on slab rates (e.g., different price per unit for different ranges).

```python
class ElectricityBill:
    def __init__(self, name, units):
        self.name = name
        self.units = units

    def calculate_bill(self):
        units = self.units

        if units <= 100:
            bill = units * 5
        elif units <= 300:
            bill = 100 * 5 + (units - 100) * 8
        elif units <= 700:
            bill = 100 * 5 + 200 * 8 + (units - 300) * 12
        else:
            bill = 100 * 5 + 200 * 8 + 400 * 12 + (units - 700) * 18

        return bill


c1 = ElectricityBill("sohail", 250)
c2 = ElectricityBill("malik mubasir", 800)

print(c1.name, "Bill:", c1.calculate_bill())  
print(c2.name, "Bill:", c2.calculate_bill()) 
```

4. Design a class 'Vehicle' and create subclasses 'Car' and 'Bike'. Override a method
'start_engine()' differently in each subclass to demonstrate polymorphism.

```python
class Vehicle:
    def __init__(self, name):
        self.name = name

    def start_engine(self):
        print(f"{self.name}: Engine starting...")


class Car(Vehicle):
    def start_engine(self): 
        print(f"{self.name} (Car): Vroom Vroom! Engine started with key.")


class Bike(Vehicle):
    def start_engine(self): 
        print(f"{self.name} (Bike): Brrr! Engine started with kick.")

v1 = Car("Toyota")
v2 = Bike("Honda")

v1.start_engine()  
v2.start_engine() 


for vehicle in [v1, v2]:
    vehicle.start_engine()
```

5. Create a class 'Employee' with basic salary. Add a method to calculate total salary including
bonus (10%) and tax deduction (5%).

```python
class Employee:
    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = basic_salary

    def total_salary(self):
        bonus = self.basic_salary * 0.10
        tax = self.basic_salary * 0.05
        total = self.basic_salary + bonus - tax
        return total


e1 = Employee("Ali", 50000)

print(f"Basic Salary: {e1.basic_salary}")
print(f"Total Salary after bonus & tax: {e1.total_salary()}")
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

c1 = Circle(5)
r1 = Rectangle(4, 6)

print("Circle Area:", c1.area())      
print("Rectangle Area:", r1.area())   
```

7. Create a class 'HospitalPatient' with attributes name, age, and disease. Add a method to
determine priority (Emergency if age > 60 or serious disease).

```python
class HospitalPatient:
    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease

    def priority(self):
        serious_diseases = ["heart attack", "stroke", "covid", "cancer"]

        if self.age > 60 or self.disease.lower() in serious_diseases:
            return "Emergency"
        else:
            return "Normal"


p1 = HospitalPatient("salamat", 65, "fever")
p2 = HospitalPatient("jamil", 30, "Heart Attack")
p3 = HospitalPatient("bilal", 25, "cold")

print(p1.name, "-", p1.priority()) 
print(p2.name, "-", p2.priority())  
print(p3.name, "-", p3.priority())  
```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []  

    def add_item(self, name, price):
        self.items.append((name, price))
        print(f"{name} added (Rs {price})")

    def total_bill(self):
        total = sum(price for name, price in self.items)

        if total > 5000:
            discount = total * 0.10  
            total = total - discount
            print(f"Discount Applied: Rs {discount}")

        return total


cart = ShoppingCart()
cart.add_item("Shoes", 3000)
cart.add_item("Shirt", 2500)
cart.add_item("Watch", 1000)

print("Total Bill:", cart.total_bill())
```

9. Create a class 'MobilePhone' with private attribute battery percentage. Use getter and setter
methods to update battery with validation (0–100 only).

```python
class MobilePhone:
    def __init__(self, model, battery):
        self.model = model
        self.__battery = battery  # private attribute

    
    def get_battery(self):
        return self.__battery

    def set_battery(self, value):
        if 0 <= value <= 100:
            self.__battery = value
            print(f"Battery updated to {value}%")
        else:
            print("Invalid value! Battery must be between 0 and 100.")


phone = MobilePhone("iPhone", 50)

print("Battery:", phone.get_battery())  

phone.set_battery(80)    
print("Battery:", phone.get_battery())  

phone.set_battery(120)   

```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Library:
    def __init__(self):
        self.books = {}

    def add_book(self, name, quantity):
        self.books[name] = self.books.get(name, 0) + quantity
        print(f"{quantity} copies of '{name}' added.")

    def borrow_book(self, name):
        if name in self.books and self.books[name] > 0:
            self.books[name] -= 1
            print(f"'{name}' borrowed successfully.")
        else:
            print(f"'{name}' is not available.")

    def return_book(self, name):
        if name in self.books:
            self.books[name] += 1
            print(f"'{name}' returned successfully.")
        else:
            print(f"'{name}' does not belong to this library.")


lib = Library()
lib.add_book("Python Basics", 2)

lib.borrow_book("Python Basics")  
lib.borrow_book("Python Basics") 
lib.borrow_book("Python Basics")  

lib.return_book("Python Basics") 
```
